```python
import numpy as np
```

# Пункт 1

```python
df_sample = np.genfromtxt("sample.csv", delimiter=",")
```

# Пункт 2

```python
data_mean = np.mean(df_sample[:], axis=1)
print(data_mean)
```

```python
data_std = np.std(df_sample[:], axis=1)
print(data_std)
```

## Если вдруг, кому-то захотелось построчно

```python
for index, row in enumerate(df_sample):
    print(f'Row{index}\tAVG={row.mean()}\tSTD={row.std()}')
```

# Пункт 3

```python
renamed_rows_temp = np.array(range(2, len(df_sample)+2))
renamed_rows_df = np.vstack([renamed_rows_temp, data_mean, data_std]).T
renamed_rows_df
```

# Пункт 4

```python
std_less_250_df = renamed_rows_df[renamed_rows_df[..., 2] < 250]
```

```python
np.savetxt('output.txt', std_less_250_df, delimiter='\t', fmt='%.2f')
```
