```python
import matplotlib.pyplot as plt

i = 0

for key, value in df_pos: # для каждой группы делаем вычисления
    print(key[0])
    current_df = pd.DataFrame(value)
#     current_df = df[df.POS == key[0]]
#     print(current_df)
    current_df.WORD = current_df.WORD.apply(len)
#     print(current_df)
#     print(current_df.WORD.rolling(window=3).mean())
    current_df['ROLLING_MEAN'] = current_df.WORD.rolling(window=3).mean() # вычисление скользящего среднего
    fig1 = current_df.plot(x='WORDNO', y='ROLLING_MEAN', label=f'POS={key[0]}', ylabel='ROLLING_MEAN')
    fig2 = fig1.get_figure()
    plt.show()
    fig2.savefig(f'C:\\Users\\Dmeezz\\Documents\\spbu\\Python\\pandas_pas\\plot\\graphs_{i}.png')
    plt.clf()
    current_df = current_df.style.set_caption(f'POS tag set as given {key[0]}') # ставим название таблицы
    current_df.to_html(f'C:\\Users\\Dmeezz\\Documents\\spbu\\Python\\pandas_pas\\table\\table_{i}.html', table_id = key[0])
    i += 1
    
# pd.DataFrame(df_pos)
```
